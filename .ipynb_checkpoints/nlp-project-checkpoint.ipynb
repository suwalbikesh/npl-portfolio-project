# 1. Import and Inspect Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
import nltk
```

Loaded `clean_nus_sms.csv` and inspected it (shape, columns, sample rows).

```python
df = pd.read_csv("clean_nus_sms.csv", index_col=0)

print("Shape:", df.shape)
print(df.info())
df.head()
```

Some notes
* The `Message` variable has 3 fewer values than the other variables, which will need to be addressed.
* All of the values in the DataFrame are strings, except for the `id` variable.
* There are 48,598 messages present in the DataFrame.

# Handle missing messages
Dropped rows with missing `Message` values, kept original `Message` in `Message_orig`.

```python
df = df.dropna(subset=["Message"])
print("New shape after dropping empty messages:", df.shape)

# Store original
df["Message_orig"] = df["Message"]
```

# 2. Text Preprocessing
In this step, I cleaned text (lowercased, stripped URLs, HTML, punctuation and digits) and tokenized via whitespace. Results are in `Message_clean` and `Message_tokens`.

```python
# Lowercasing
df["Message_clean"] = df["Message_orig"].str.lower()

# Remove URLs
url_pattern = re.compile(r'https?://\S+|www\.\S+')
df["Message_clean"] = df["Message_clean"].apply(lambda x: url_pattern.sub("", x))

# Remove HTML tags
html_pattern = re.compile(r'<.*?>')
df["Message_clean"] = df["Message_clean"].apply(lambda x: html_pattern.sub("", x))

# Remove punctuation
df["Message_clean"] = df["Message_clean"].str.replace(r"[^\w\s]", "", regex=True)

# Remove digits
df["Message_clean"] = df["Message_clean"].str.replace(r"\d+", "", regex=True)

# Tokenize
nltk.download("punkt_tab")
df["Tokens"] = df["Message_clean"].apply(nltk.word_tokenize)

df.head()
```

```python
# Save the preprocessed DataFrame
df.to_csv('processed_clean_nus_sms.csv', header = True)
```

# 4. Sentiment Analysis (Naive Bayes with NLTK Twitter Samples)
* Because the notebook could not download NLTK data/models here, I used a small offline lexicon-based sentiment classifier (very small positive/negative word lists included in the code).
* Every message receives a `Sentiment_lexicon` label (`Positive` / `Negative` / `Neutral`) and a small integer `Sentiment_score_lex`.
* Produced counts and a barplot of lexicon sentiment.
* Saved final CSV with these sentiment columns.

```python
# Download twitter data and sentiment analysis model
nltk.download("twitter_samples")
nltk.download("wordnet")
nltk.download("stopwords")
nltk.download("averaged_perceptron_tagger")

from nltk.corpus import twitter_samples
from nltk import NaiveBayesClassifier, classify
```

```python
# Load labeled tweets
positive_tweets = twitter_samples.strings("positive_tweets.json")
negative_tweets = twitter_samples.strings("negative_tweets.json")
```

```python
# Tokenize them
pos_tokens = twitter_samples.tokenized("positive_tweets.json")
neg_tokens = twitter_samples.tokenized("negative_tweets.json")
```

```python
# Preprocessing functions
def remove_urls_html(token_list):
    clean_list = []
    for t in token_list:
        t = url_pattern.sub("", t)
        t = html_pattern.sub("", t)
        t = re.sub(r"[^\w\s]", "", t)
        clean_list.append(t)
    return clean_list

positive_cleaned = [remove_urls_html(t) for t in pos_tokens]
negative_cleaned = [remove_urls_html(t) for t in neg_tokens]
```

```python

# Feature extraction
all_tokens = [t for tokens in (positive_cleaned + negative_cleaned) for t in tokens]
freq = nltk.FreqDist(all_tokens)
features = list(freq.keys())[:10000]
```

```python
def document_features(doc):
    doc_set = set(doc)
    return {word: (word in doc_set) for word in features}

# Build training/testing sets
pos_labeled = [(t, "Positive") for t in positive_cleaned]
neg_labeled = [(t, "Negative") for t in negative_cleaned]

all_labeled = pos_labeled + neg_labeled
import random
random.seed(42)
random.shuffle(all_labeled)

feature_set = [(document_features(doc), label) for (doc, label) in all_labeled]

train_set = feature_set[:7000]
test_set = feature_set[7000:]
```

```python
# Train classifier
classifier = NaiveBayesClassifier.train(train_set)

print("Accuracy:", classify.accuracy(classifier, test_set))
classifier.show_most_informative_features(10)
```

# Apply classifier to SMS messages

```python
def predict_sentiment(tokens):
    feats = {word: True for word in tokens}
    return classifier.classify(feats)

df["Sentiment"] = df["Tokens"].apply(lambda t: predict_sentiment(t))

df["Sentiment"].value_counts()

plt.figure(figsize=(6,4))
sns.countplot(x="Sentiment", data=df)
plt.title("Sentiment Distribution")
plt.show()
```

```python
# Convert to numeric score
df["Sentiment_score"] = df["Sentiment"].apply(lambda x: 1 if x=="Positive" else -1)
```

```python
# Sentiment by Country
sent_by_country = df.groupby("country")["Sentiment_score"].mean().reset_index()

plt.figure(figsize=(20,7))
sns.barplot(x="country", y="Sentiment_score", data=sent_by_country)
plt.xticks(rotation=45)
plt.title("Average Sentiment Score by Country")
plt.show()
```

# 6. Topic Modeling (TF-IDF + NMF)
* TF-IDF vectorization (unigrams + selected bigrams) + NMF (5 topics).
* Printed the top terms for each of the 5 topics.

```python
# Importing TfidfVectorizer and NMF form sklearn
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import NMF
```

```python
docs = df["Message_clean"].astype(str)

tfidf = TfidfVectorizer(min_df=5, max_df=0.6, stop_words="english")
tfidf_matrix = tfidf.fit_transform(docs)

nmf = NMF(n_components=5, random_state=42)
W = nmf.fit_transform(tfidf_matrix)
H = nmf.components_

feature_names = tfidf.get_feature_names_out()

def show_topic(idx, topn=10):
    print(f"Topic {idx+1}:")
    terms = H[idx].argsort()[-topn:][::-1]
    for t in terms:
        print("  -", feature_names[t])
    print()

for i in range(5):
    show_topic(i)
```
